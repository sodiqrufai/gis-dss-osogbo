```python
import numpy as np
from shapely.geometry import box

# Define Osogbo bounding box (same as before)
west, south, east, north = 4.52, 7.74, 4.60, 7.80

# Grid cell size in degrees (~0.005 deg ≈ 500m at this latitude)
cell_size = 0.005

# Generate grid cells
cells = []
x = west
while x < east:
    y = south
    while y < north:
        cells.append(box(x, y, x + cell_size, y + cell_size))
        y += cell_size
    x += cell_size

grid = gpd.GeoDataFrame({'zone_type': 'commercial_candidate'}, geometry=cells, crs="EPSG:4326")
print(f"Total grid cells: {len(grid)}")

# Quick visualization
fig, ax = plt.subplots(figsize=(10, 10))
roads_check.plot(ax=ax, color='gray', linewidth=0.3)
grid.boundary.plot(ax=ax, color='blue', linewidth=0.5)
plt.title(f"Osogbo Analysis Grid: {len(grid)} cells (~500m x 500m)")
plt.show()
```

```python
# Load zones and roads back from PostGIS
zones = gpd.read_postgis("SELECT * FROM land_use_zone", engine, geom_col='geom')
roads = gpd.read_postgis("SELECT * FROM road_network", engine, geom_col='geom')

print(f"Zones: {len(zones)}, Roads: {len(roads)}")

# Define "major roads" - primary, trunk, secondary
major_roads = roads[roads['road_class'].isin(['primary', 'trunk', 'secondary', 'primary_link', 'trunk_link', 'secondary_link'])]
print(f"Major roads: {len(major_roads)}")
```

```python
# IMPORTANT: lat/lon distances aren't in meters. We need to reproject to a metric CRS.
# UTM Zone 31N (EPSG:32631) is appropriate for this longitude range in Nigeria

zones_m = zones.to_crs(epsg=32631)
major_roads_m = major_roads.to_crs(epsg=32631)

# Calculate centroid of each zone
zones_m['centroid'] = zones_m.geometry.centroid

# Merge all major roads into a single geometry for distance calculation
major_roads_union = major_roads_m.union_all()

# Calculate distance from each zone centroid to nearest major road (in meters)
zones_m['dist_major_road'] = zones_m['centroid'].apply(lambda c: c.distance(major_roads_union))

print(zones_m[['zone_id', 'dist_major_road']].describe())
zones_m[['zone_id', 'dist_major_road']].head(10)
```

```python
# Reproject all roads (not just major) to metric CRS
roads_all_m = roads.to_crs(epsg=32631)

# For each zone, calculate total road length within it (in meters)
road_density = []
for idx, zone in zones_m.iterrows():
    # Find roads that intersect this zone
    clipped = roads_all_m.geometry.intersection(zone.geometry)
    total_length = clipped.length.sum()
    road_density.append(total_length)

zones_m['road_density'] = road_density

print(zones_m[['zone_id', 'road_density']].describe())
zones_m[['zone_id', 'road_density']].head(10)
```

```python
roads_all_m = roads.to_crs(epsg=32631)

road_density = []
for idx, zone in zones_m.iterrows():
    clipped = roads_all_m.geometry.intersection(zone.geom)
    total_length = clipped.length.sum()
    road_density.append(total_length)

zones_m['road_density'] = road_density

print(zones_m[['zone_id', 'road_density']].describe())
zones_m[['zone_id', 'road_density']].head(10)
```

```python
pois_m = pois_check.to_crs(epsg=32631)

# Spatial join: count POIs within each zone
zones_for_join = zones_m[['zone_id', 'geom']].set_geometry('geom')

poi_counts = gpd.sjoin(pois_m, zones_for_join, how='left', predicate='within')
poi_count_per_zone = poi_counts.groupby('zone_id').size().reset_index(name='poi_count')

zones_m = zones_m.merge(poi_count_per_zone, on='zone_id', how='left')
zones_m['poi_count'] = zones_m['poi_count'].fillna(0).astype(int)

print(zones_m[['zone_id', 'poi_count']].describe())
print(f"\nZones with at least 1 POI: {(zones_m['poi_count'] > 0).sum()} out of {len(zones_m)}")
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

zones_m['dist_norm'] = scaler.fit_transform(zones_m[['dist_major_road']])
zones_m['dist_score'] = 1 - zones_m['dist_norm']

zones_m['density_score'] = scaler.fit_transform(zones_m[['road_density']])

zones_m['poi_score'] = scaler.fit_transform(zones_m[['poi_count']])

ahp_weights = pd.read_sql("SELECT criteria_name, weight FROM ahp_criteria", engine)
print(ahp_weights)
```

```python
# Rescale our 3 available weights to sum to 1.0
w_dist = 0.15
w_density = 0.05
w_poi = 0.10
total_w = w_dist + w_density + w_poi  # 0.30

w_dist_norm = w_dist / total_w
w_density_norm = w_density / total_w
w_poi_norm = w_poi / total_w

print(f"Normalized weights: dist={w_dist_norm:.3f}, density={w_density_norm:.3f}, poi={w_poi_norm:.3f}")

# Compute composite AHP suitability score (0-100 scale)
zones_m['ahp_score'] = (
    zones_m['dist_score'] * w_dist_norm +
    zones_m['density_score'] * w_density_norm +
    zones_m['poi_score'] * w_poi_norm
) * 100

print(zones_m[['zone_id', 'ahp_score']].describe())
zones_m.nlargest(10, 'ahp_score')[['zone_id', 'ahp_score', 'dist_major_road', 'road_density', 'poi_count']]
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import MinMaxScaler
import numpy as np
import pickle
import os

# ── Build feature matrix ─────────────────────────────────────────────────────
zones_features = gpd.read_postgis(
    "SELECT zone_id, geom FROM land_use_zone", engine, geom_col='geom'
)
zones_features = zones_features.to_crs(epsg=32631)

# Attach building features
zones_features = zones_features.merge(
    zones[['zone_id', 'building_count', 'building_density_score']], on='zone_id', how='left'
)

# Distance to major roads
major_roads = gpd.read_postgis(
    "SELECT geom FROM road_network WHERE road_class IN ('primary','trunk','secondary','primary_link','trunk_link')",
    engine, geom_col='geom'
).to_crs(epsg=32631)
major_roads_union = major_roads.union_all()

zones_features['centroid'] = zones_features['geom'].centroid
zones_features['dist_major_road'] = zones_features['centroid'].apply(
    lambda c: c.distance(major_roads_union)
)

# Road density — use zone.geom not zone.geometry
all_roads = gpd.read_postgis(
    "SELECT geom FROM road_network", engine, geom_col='geom'
).to_crs(epsg=32631)

road_density = []
for idx, zone in zones_features.iterrows():
    clipped = all_roads['geom'].intersection(zone.geom)
    road_density.append(clipped.length.sum())
zones_features['road_density'] = road_density

# POI count
pois_gdf = gpd.read_postgis(
    "SELECT geom FROM business_poi", engine, geom_col='geom'
).to_crs(epsg=32631)
zones_for_join = zones_features[['zone_id', 'geom']].set_geometry('geom')
poi_join = gpd.sjoin(pois_gdf, zones_for_join, how='left', predicate='within')
poi_count = poi_join.groupby('zone_id').size().reset_index(name='poi_count')
zones_features = zones_features.merge(poi_count, on='zone_id', how='left')
zones_features['poi_count'] = zones_features['poi_count'].fillna(0).astype(int)

print("Features computed!")
print(zones_features[['zone_id', 'dist_major_road', 'road_density', 'building_count', 'poi_count']].describe())
```

```python
# ── Load target variable (composite AHP scores from database) ────────────────
scores_df = pd.read_sql(
    "SELECT zone_id, composite_score FROM suitability_score", engine
)
zones_features = zones_features.merge(scores_df, on='zone_id', how='left')

# ── Define feature matrix X and target y ─────────────────────────────────────
feature_cols = ['dist_major_road', 'road_density', 'building_count', 'poi_count']
X = zones_features[feature_cols].copy()
y = zones_features['composite_score'].copy()

print(f"Dataset: {len(X)} zones, {len(feature_cols)} features")
print(f"Target range: {y.min():.2f} to {y.max():.2f}")

# ── Normalize features ────────────────────────────────────────────────────────
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=feature_cols)

# Invert dist_major_road (closer = better)
X_scaled['dist_major_road'] = 1 - X_scaled['dist_major_road']

# ── Train/test split ──────────────────────────────────────────────────────────
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
print(f"Train: {len(X_train)}, Test: {len(X_test)}")

# ── Train Random Forest ───────────────────────────────────────────────────────
rf_model = RandomForestRegressor(
    n_estimators=200,
    max_depth=8,
    min_samples_leaf=2,
    random_state=42,
    n_jobs=-1
)
rf_model.fit(X_train, y_train)

# ── Evaluate ──────────────────────────────────────────────────────────────────
y_pred = rf_model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

# Cross-validation
cv_scores = cross_val_score(rf_model, X_scaled, y, cv=5, scoring='r2')

print(f"\nModel Performance:")
print(f"  RMSE:              {rmse:.4f}")
print(f"  R² (test):         {r2:.4f}")
print(f"  R² (5-fold CV):    {cv_scores.mean():.4f} (+/- {cv_scores.std():.4f})")

# Feature importance
importance_df = pd.DataFrame({
    'feature': feature_cols,
    'importance': rf_model.feature_importances_
}).sort_values('importance', ascending=False)
print(f"\nFeature Importances:")
print(importance_df.to_string(index=False))
```

```python
# ── Generate ML scores for ALL 131 zones ─────────────────────────────────────
ml_predictions = rf_model.predict(X_scaled)

zones_features['ml_score'] = ml_predictions

print(f"ML Score range: {ml_predictions.min():.2f} to {ml_predictions.max():.2f}")
print(f"ML Score mean:  {ml_predictions.mean():.2f}")

# ── Compute final hybrid composite score ─────────────────────────────────────
# Combine AHP (interpretable, expert-driven) + ML (data-driven) 50/50
zones_features['hybrid_score'] = (
    zones_features['composite_score'] * 0.5 +
    zones_features['ml_score'] * 0.5
)

print(f"\nHybrid Score range: {zones_features['hybrid_score'].min():.2f} to {zones_features['hybrid_score'].max():.2f}")
print(f"Hybrid Score mean:  {zones_features['hybrid_score'].mean():.2f}")

# Top 10 by hybrid score
top10_hybrid = zones_features.nlargest(10, 'hybrid_score')[
    ['zone_id', 'composite_score', 'ml_score', 'hybrid_score']
]
print(f"\nTop 10 Zones (Hybrid Score):")
print(top10_hybrid.to_string(index=False))
```
